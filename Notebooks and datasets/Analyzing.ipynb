```python
import pandas as pd
```

```python
ruas = pd.read_csv('ruaslx.csv')
```

```python
ruas.columns
```

```python
ruas['FREGUESIAS'] = ruas['FREGUESIAS'].str.replace(r'\s*\(Nova Freguesia\)', '', regex=True)
```

```python

ruas['FREGUESIAS'] = ruas['FREGUESIAS'].str.split(',')

ruas_explodido = ruas.explode('FREGUESIAS')

ruas_explodido['FREGUESIAS'] = ruas_explodido['FREGUESIAS'].str.strip()

contagem = ruas_explodido.groupby('FREGUESIAS')['rua_padronizada'].nunique().reset_index()

contagem.columns = ['FREGUESIA', 'Qtd_Ruas']

print(contagem)
```

```python
ruas['rua_padronizada'] = ruas['rua_padronizada'].str.upper()
ruas
```

```python
ruas.to_csv("padroesruas.csv", index=False)
```

```python
al = pd.read_csv('Final.csv')
```

```python
total_camas = al['NrCamas'].sum()
total_utentes = al['NrUtentes'].sum()

print("Total de camas:", total_camas)
print("Total de utentes:", total_utentes)
```

```python
totais_por_freguesia = al.groupby('Freguesia')[['NrCamas', 'NrUtentes']].sum()
print(totais_por_freguesia)
```

```python
al
```

```python
al['RuaMatch'] = al['RuaMatch'].str.strip()
al['Freguesia'] = al['Freguesia'].str.strip()

al_contagem = al.groupby('Freguesia')['RuaMatch'].nunique().reset_index()

al_contagem.columns = ['FREGUESIA', 'Qtd_Ruas_AL']
```

```python
comparacao = contagem.merge(al_contagem, on='FREGUESIA', how='left')

comparacao['Qtd_Ruas_AL'] = comparacao['Qtd_Ruas_AL'].fillna(0).astype(int)

comparacao['Proporcao_AL_vs_Oficial'] = (comparacao['Qtd_Ruas_AL'] / comparacao['Qtd_Ruas']) * 100

comparacao['Proporcao_AL_vs_Oficial'] = comparacao['Proporcao_AL_vs_Oficial'].round(2)

print(comparacao)

```

```python
top_5 = comparacao.sort_values(by='Proporcao_AL_vs_Oficial', ascending=False).head(5)
print(top_5)
```

```python

contagem_por_rua = al['RuaMatch'].value_counts().reset_index()

contagem_por_rua.columns = ['Rua', 'NrAlojamentos']

contagem_por_rua = contagem_por_rua.sort_values(by='NrAlojamentos', ascending=False)

print(contagem_por_rua)
```

```python
import altair as alt
import pandas as pd

comparacao['Qtd_Ruas_AL'] = comparacao['Qtd_Ruas_AL'].fillna(0).astype(int)
comparacao['Proporcao_AL_vs_Oficial'] = (comparacao['Qtd_Ruas_AL'] / comparacao['Qtd_Ruas']) * 100
comparacao['Proporcao_AL_vs_Oficial'] = comparacao['Proporcao_AL_vs_Oficial'].round(2)

top_n = comparacao.sort_values(by='Proporcao_AL_vs_Oficial', ascending=False).head(25)

top_n['FREGUESIA_LABEL'] = top_n['FREGUESIA'] + '\n' + top_n['Proporcao_AL_vs_Oficial'].astype(str) + '%'

ordem_freguesias = top_n['FREGUESIA_LABEL'].tolist()

dados_retangulos = []

for _, row in top_n.iterrows():
    freguesia_label = row['FREGUESIA_LABEL']
    proporcao = int(round(row['Proporcao_AL_vs_Oficial']))

    df = pd.DataFrame({'id': range(100)})
    df['x'] = df['id'] % 10
    df['y'] = 9 - (df['id'] // 10)
    df['preenchido'] = df['id'] < proporcao
    df['FREGUESIA_LABEL'] = freguesia_label
    df['Proporcao'] = proporcao
    dados_retangulos.append(df)

df_final = pd.concat(dados_retangulos)

base = alt.Chart(df_final).mark_rect(
    stroke='white',
    strokeWidth=0.5
).encode(
    x=alt.X('x:O', axis=None),
    y=alt.Y('y:O', axis=None),
    color=alt.condition(
        'datum.preenchido',
        alt.value('#84A4BF'),
        alt.value('#e0e0e0')
    )
).properties(
    width=60,
    height=60
)

grafico = base.facet(
    facet=alt.Facet('FREGUESIA_LABEL:N', sort=ordem_freguesias),
    columns=5,
    title='Proporção de AL por Freguesia (Top 25)'
)

grafico

```
